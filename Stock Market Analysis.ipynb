# Comparing Silver and Gold's closing price for the period 2011-2013 using Resampling

#From the above graphs it was clear that prices of gold and silver were high between 2011-2013. Thus, now using resmapling we can focus only on that period of time and see the resulting trend.

```python
plt.figure(figsize=(10,5))
gold_pd =gold.loc[pd.Timestamp('2011-01-01'):pd.Timestamp('2013-12-31')]
plt.plot(gold_pd.index, gold_pd['Adj Close'],'ro')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xlabel('Year')
plt.ylabel('Closing Price')         
plt.title('Closing price trend for year 2011-13') 
plt.xticks(rotation=90)
```

#thus, the price of gold was high in months of august, septempber and october in 2011 and then again in november and december. This was probably because of the festival season in india.

#The price was high in the first half of 2012 as well and fell down immediately after. Same reason of festivals could be applied to the high price at the end of 2012. 

#Since then the prices have been comparatively lower.

```python
plt.figure(figsize=(10,5))
silver_pd=silver.loc[pd.Timestamp('2011-01-01'):pd.Timestamp('2013-12-31')]
plt.plot(silver_pd.index, silver_pd['Adj Close'],'ro', color='blue')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xlabel('Year')
plt.ylabel('Closing price')
plt.title('Closing price Trends for year 2011-2013')
plt.xticks(rotation=90)
```
